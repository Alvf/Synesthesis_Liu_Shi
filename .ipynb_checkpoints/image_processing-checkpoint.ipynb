```python
# import necessary packages
from PIL import Image
import cv2
import matplotlib.pyplot as plt
import scipy
import numpy as np
import colorsys
import collections
from sklearn.cluster import KMeans
import skimage.measure
import pandas as pd
import math
import random

# define global variables
BASE_DIM = 300
```

```python
# method to read in and crop image to square
def crop_pic(img_path, new_sq = (BASE_DIM, BASE_DIM)):
    
    # open image
    im = Image.open(img_path)

    # get width and height, dim of square image
    w, h = im.size
    sq = np.min([w, h])
    sq_in = [w, h].index(sq)

    # crop image to square
    if sq_in == 0:
        im_crop = im.crop((0,
                           (h / 2) - (int)(sq / 2),
                           w, 
                           (h / 2) + (int)(sq / 2)))
    else:
        im_crop = im.crop(((w / 2) - (int)(sq / 2),
                           0,
                           (w / 2) + (int)(sq / 2), 
                           h))

    # resize cropped image
    im_rs = im_crop.resize(new_sq)
    return im_rs
    
# method to get RGB color channels
def get_rgb(im_rs):
    
    # get RGB color channels
    im_ar = np.array(im_rs)
    r, g, b = im_ar[:, :, 0], im_ar[:, :, 1], im_ar[:, :, 2]
    
    # scale RGB to [0, 1]
    r_sc, g_sc, b_sc = r / 255., g / 255., b / 255.
    
    return r_sc, g_sc, b_sc

# method to get HSV from RGB
def get_hsv(r_sc, g_sc, b_sc):
    
    # create empty arrays
    h = np.zeros((BASE_DIM, BASE_DIM))
    s = np.zeros((BASE_DIM, BASE_DIM))
    v = np.zeros((BASE_DIM, BASE_DIM))
    
    # compute hsv for each pixel based on normalized RGB
    for i in range(BASE_DIM):
        for j in range(BASE_DIM):
            temp_hsv = colorsys.rgb_to_hsv(r_sc[i, j], b_sc[i, j], g_sc[i, j])
            h[i][j] = temp_hsv[0]
            s[i][j] = temp_hsv[1]
            v[i][j] = temp_hsv[2]
    
    # return HSV
    return h, s, v

# method to find 'clusters' of hues in the image using k-means
def get_clusts(h, n_clusters = 3, red_scale = 20):
    
    # implement k-means algorithm to cluster vectorized pixels by hue
    x = h
    km = KMeans(n_clusters)
    km.fit(x.reshape(-1,1))
    
    # debug: plot histogram
    #plt.hist(list(km.labels_), bins = 80)
    
    # reshape kmeans back into 2D array
    test_ar = np.array(km.labels_).reshape((300, 300))
    
    # pool kmeans into lower dimension to get rid of noise
    test_ar_red = skimage.measure.block_reduce(test_ar, (red_scale, red_scale), np.median)
    
    # upscale block reduced kmeans clusterings
    test_ar_red_rs = np.zeros((BASE_DIM, BASE_DIM))
    for i in range(BASE_DIM):
        for j in range(BASE_DIM):
            i_red = (int)(np.floor(i / red_scale))
            j_red = (int)(np.floor(j / red_scale))
            test_ar_red_rs[i][j] = test_ar_red[i_red][j_red]
    
    # return BASE_DIM x BASE_DIM 2d array containing pixel clustering results
    return test_ar_red_rs

# get average h, s range, v range, pan, an reverb variables based on pixel clusterings
# average h is a hue value, to be converted back into RGB ([0, 100]) later
# s range in [0, 1] describes color saturation, to be converted to filter cutoff
# v range in indices [0, 8] of {-12, -8, -7, -5, 0, +4, +5, +7, +12} describes color value, to be converted to frequency / note modulation
# pan roughly describes the 'x' position of where the centroid of the pixel cluster is, [0, 1], describes L-R panning
# pan roughly describes the 'y' position of where the centroid of the pixel cluster is, [0, 1], the higher the wetter
def get_hsv_quants(h, s, v, pix_ar):
    
    # create empty lists
    avg_h, low_s, high_s, low_v, high_v, pan, reverb = [], [], [], [], [], [], []
    
    # get counts of clusters
    cluster_labs, cluster_cnts = np.unique(pix_ar, return_counts=True)
    
    # debug, print cluster labels
    #print(cluster_labs)
    
    # for each cluster...
    for lab in cluster_labs:
        hue_total = 0
        temp_low_s = float("inf")
        temp_high_s = 0
        temp_low_v = float("inf")
        temp_high_v = 0
        temp_right_pan = pix_ar.shape[0]
        temp_left_pan = 0
        temp_high_rev = pix_ar.shape[0]
        temp_low_rev = 0
        for i in range(pix_ar.shape[0]):
            for j in range(pix_ar.shape[1]):
                if (pix_ar[i][j] == lab):
                    hue_total += h[i][j] # add to hue if within cluster
                    if s[i][j] < temp_low_s: temp_low_s = s[i][j]
                    if s[i][j] > temp_high_s: temp_high_s = s[i][j]
                    if v[i][j] < temp_low_v: temp_low_v = v[i][j]
                    if v[i][j] > temp_high_v: temp_high_v = v[i][j]
                    if j < temp_right_pan: temp_right_pan = j
                    if j > temp_left_pan: temp_left_pan = j
                    if i < temp_high_rev: temp_high_rev = i
                    if i > temp_low_rev: temp_low_rev = i
        
        avg_h.append(hue_total / cluster_cnts[list(cluster_labs).index(lab)]) # get average hue
        low_s.append(temp_low_s) # append low saturation
        high_s.append(temp_high_s) # append high saturation
        low_v.append(temp_low_v) # append low value
        high_v.append(temp_high_v) # apped high value
        pan.append(((temp_right_pan + temp_left_pan) / 2) / pix_ar.shape[0])
        reverb.append(np.abs(((temp_high_rev + temp_low_rev) / 2) - pix_ar.shape[0]) / pix_ar.shape[0])
    
    # convert v to integer indices
    low_v_ind = [i * 8 for i in low_v]
    high_v_ind = [i * 8 for i in high_v]
    low_v_ind = [round(j) for j in low_v_ind]
    high_v_ind = [round(j) for j in high_v_ind]
    
    # convert list of hues to RGB
    avg_r, avg_g, avg_b = [], [], []
    
    # convert hues to RGB, with full saturation and value
    for hue in avg_h:
        temp_color = colorsys.hsv_to_rgb(hue, 1, 1)
        avg_r.append(temp_color[0])
        avg_g.append(temp_color[1])
        avg_b.append(temp_color[2])
    
    # return lists containing output values
    return avg_r, avg_g, avg_b, low_s, high_s, low_v_ind, high_v_ind, pan, reverb
```

```python
# method to convert image to quantized sonic values
def quantize_image(filename, n_clusters = 3):
    
    # crop image
    im_rs = crop_pic(filename)
    
    # get RGB channels
    r, g, b = get_rgb(im_rs)
    
    # get HSV from RGB channels
    h, s, v = get_hsv(r, g, b)
    
    # get pixel clusters (by hue)
    pix_ar = get_clusts(h, n_clusters)
    
    # return R, G, B, saturation range, value range, pan, and reverb
    avg_r, avg_g, avg_b, low_s, high_s, low_v_ind, high_v_ind, pan, reverb = get_hsv_quants(h, s, v, pix_ar)
    
    # create and populate dataframe
    df_exp1 = pd.DataFrame()
    df_exp1['r'] = avg_r
    df_exp1['g'] = avg_g
    df_exp1['b'] = avg_b
    df_exp1['low_s'] = low_s
    df_exp1['high_s'] = high_s
    df_exp1['low_v'] = low_v_ind
    df_exp1['high_v'] = high_v_ind
    df_exp1['pan'] = pan
    df_exp1['reverb'] = reverb
    
    # export to csv
    df_exp1.to_csv("output_1.csv", index = False)
    
    # get Shannon entropy for R, G, B color channels and grayscale image   
    r_entr = (int)(np.floor(skimage.measure.shannon_entropy(r)))
    g_entr = (int)(np.floor(skimage.measure.shannon_entropy(g)))
    b_entr = (int)(np.floor(skimage.measure.shannon_entropy(b)))
    im_gs = 0.2989 * r + 0.587 * g + 0.114 * b
    gr_entr = (int)(np.floor(skimage.measure.shannon_entropy(im_gs)))
    
    # create and populate dataframe (entropy)
    df_exp2 = pd.DataFrame()
    df_exp2['grayscale'] = [gr_entr]
    df_exp2['r'] = [r_entr]
    df_exp2['g'] = [g_entr]
    df_exp2['b'] = [b_entr]
    
    # explort to csv
    df_exp2.to_csv("output_2.csv", index= False)
```

```python
# fft chords
    # low freq --> more pure block (relaxing sounds)
    # high freq --> noise (frantic chord progressions)

# method to get sample spectra from 2D FFT of input image RGB channels
def get_spectr(r_img, g_img, b_img, pad = 5, n_partials = 10, max_freq = 20000, mult_pow = 0.45):
    
    # compute 2D FFTs
    r_ft, g_ft, b_ft = np.fft.fft2(r_img), np.fft.fft2(g_img), np.fft.fft2(b_img)
    
    # shift 2D FFTs and take log transform
    r_shift, g_shift, b_shift = np.fft.fftshift(r_ft), np.fft.fftshift(g_ft), np.fft.fftshift(b_ft)
    r_spect, g_spect, b_spect = np.log(np.abs(r_shift)), np.log(np.abs(g_shift)), np.log(np.abs(b_shift))
    
    # return diagonal values in first quadrant
    r_d, g_d, b_d = np.diagonal(r_spect), np.diagonal(g_spect), np.diagonal(b_spect)
    r_db, g_db, b_db = [i - np.max(r_d) for i in r_d], [i - np.max(g_d) for i in g_d], [i - np.max(b_d) for i in b_d]
    r_db, g_db, b_db = r_db[0: (int)(len(r_db)/2)], g_db[0: (int)(len(g_db)/2)], b_db[0: (int)(len(b_db)/2)]
    r_db, g_db, b_db = r_db[0 + pad: len(r_db) - pad], g_db[0 + pad: len(g_db) - pad], b_db[0 + pad: len(b_db) - pad]
    
    # resize spectra
    r_res = -1 * cv2.resize(np.abs(r_db), dsize = (1, max_freq), interpolation = cv2.INTER_CUBIC)
    g_res = -1 * cv2.resize(np.abs(g_db), dsize = (1, max_freq), interpolation = cv2.INTER_CUBIC)
    b_res = -1 * cv2.resize(np.abs(b_db), dsize = (1, max_freq), interpolation = cv2.INTER_CUBIC)
    
    # make spectra quieter
    r_res -= 10
    g_res -= 10
    b_res -= 10
    
    # get partials
    r_parts, g_parts, b_parts = [], [], []
    for i in range(n_partials):
        par_width = max_freq / n_partials
        r_res[i * par_width] = r_res[i * par_width] + np.abs(r_res[i * par_width])**mult_pow
        g_res[i * par_width] = g_res[i * par_width] + np.abs(g_res[i * par_width])**mult_pow
        b_res[i * par_width] = b_res[i * par_width] + np.abs(b_res[i * par_width])**mult_pow
        r_parts.append(r_res[i * par_width])
        g_parts.append(g_res[i * par_width])
        b_parts.append(b_res[i * par_width])
    
    # turn partials into flat lists
    print(r_parts)
    ###
    res = list(i[0] for i in res)
    ls_partials = list(i[0] for i in ls_partials)
    plt.plot(res)
```
